```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("class_weathering_chemical.csv")#导入数据集
dataplot = data.iloc[:,1:]
```

```python
plt.figure(figsize=[15, 10])#生成热力图,初步观察变量相关情况
sns.heatmap(pd.get_dummies(dataplot).corr(method='spearman'),cmap="Greens",mask=abs(dataplot.corr(method='spearman'))<0.255,linewidths=0.5, annot=True)#斯皮尔曼相关(适用于分类数据)
plt.show()
```
