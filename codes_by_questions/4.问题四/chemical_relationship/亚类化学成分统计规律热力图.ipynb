```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
PbBa_NOweathering = pd.read_csv(".\subclass_dataset\PbBa_NOweathering_chemical.csv")#导入四个数据集
PbBa_weathering = pd.read_csv(".\subclass_dataset\PbBa_weathering_chemical.csv")
richk_NOweathering = pd.read_csv(".\subclass_dataset\\richk_NOweathering_chemical.csv")
richk_weathering = pd.read_csv(".\subclass_dataset\\richk_weathering_chemical.csv")
```

```python
dataplot1 = PbBa_NOweathering.iloc[:,1:15]
dataplot2 = PbBa_weathering.iloc[:,1:15]
dataplot3 = richk_NOweathering.iloc[:,1:15]
dataplot4 = richk_weathering.iloc[:,1:15]
```

```python
plt.figure(figsize=[12, 9])#生成热力图,初步观察变量相关情况
sns.heatmap(dataplot1.corr(method='spearman'),mask=abs(dataplot1.corr(method='spearman'))<0.415,cmap="Greens", linewidths=0.5, annot=True)#斯皮尔曼相关(适用于分类数据)
plt.show()
```

```python
plt.figure(figsize=[12, 9])#生成热力图,初步观察变量相关情况
sns.heatmap(dataplot2.corr(method='spearman'),mask=(abs(dataplot2.corr(method='spearman'))<0.39),cmap="Greens", linewidths=0.5, annot=True)#斯皮尔曼相关(适用于分类数据)
plt.show()
```

```python
plt.figure(figsize=[12, 9])#生成热力图,初步观察变量相关情况
sns.heatmap(dataplot3.corr(method='spearman'),mask=abs(dataplot3.corr(method='spearman'))<0.587,cmap="Greens", linewidths=0.5, annot=True)#斯皮尔曼相关(适用于分类数据)
plt.show()
```

```python
plt.figure(figsize=[12, 9])#生成热力图,初步观察变量相关情况
sns.heatmap(dataplot4.corr(method='spearman'),cmap="Greens",mask=abs(dataplot4.corr(method='spearman'))<0.886,linewidths=0.5, annot=True)#斯皮尔曼相关(适用于分类数据)
plt.show()
```
